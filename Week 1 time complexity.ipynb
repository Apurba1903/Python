```python
#Reverse list

L = [1,2,3,4,5]

for i in range (0,len(L)//2):
    other = len(L) - i - 1
    temp = L[i]
    L[i] = L[other]
    L[other] = temp

print (L)

# Tme Complexity = Order of (n)
```

```python
n = 10
k = 0
for i in range (n//2,n):
    for j in range (2,n,pow(2,j)):
        k = k + n / 2

print(k)

# Tme Complexity = n Log n
```

```python
#Sum of all digit

n =  1903
sum = 0

while n>0:
    sum = sum + n%10
    n = n//10

print (sum)

# Tme Complexity = Log n
```

```python
# Fibonacci Series

def fib(n):
    if n == 1 or n == 0:
        return 1
    else:
        return fib(n-1) + fib(n-2)

# Tme Complexity = 2^n
```
